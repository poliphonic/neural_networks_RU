```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import keras
from keras.datasets import fashion_mnist
from keras.models import Sequential
from keras.layers import Dense, Flatten
from keras.utils import to_categorical

from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
# Import MINST data
mnist = keras.datasets.mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
# Normalize the images.
train_images = (train_images / 255)
test_images = (test_images / 255)

# Flatten the images.
train_images = train_images.reshape((-1, 784))
test_images = test_images.reshape((-1, 784))
```

```python
# Build the model.
def create_model(neurons, layers_num, optimizer, batches, epochs):

    layers_lst = [Dense(64, input_shape=(784,), activation='relu')]
    for i in range(layers_num):
        layers_lst += [Dense(neurons, activation='relu')]
    layers_lst +=  [Dense(10, activation='softmax')]

    model = Sequential(layers_lst)

    # Compile the model.
    model.compile(optimizer=optimizer,
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    trained = model.fit(train_images, to_categorical(train_labels),
                    epochs=epochs, batch_size=batches, verbose=0)

    train_accuracy = trained.history['accuracy'][-1]

    test_accuracy = model.evaluate(test_images, to_categorical(test_labels),
                                   verbose=0)[-1]

    epoch_seq = np.arange(1, epochs + 1)
    plt.plot(epoch_seq, trained.history['loss'], 'go-')
    plt.grid('On')

    plt.title('График функции потерь')
    plt.xlabel('Эпохи')
    plt.ylabel('Ошибка')
    plt.show()

    return round(train_accuracy, 4), round(test_accuracy, 4)


def df_add(df, neurons, layers, epochs, optimizer, batches,
           train_accuracy, test_accuracy):
    df = df._append({'Neurons number': neurons,
                     'Layers number': layers,
                     'Optimizer': optimizer,
                     'Batches': batches,
                     'Epochs': epochs,
                     'Train accuracy': train_accuracy,
                     'Test accuracy': test_accuracy}, ignore_index=True)
    df[['Neurons number', 'Layers number', 'Batches', 'Epochs']] = df[
        ['Neurons number', 'Layers number', 'Batches', 'Epochs']].astype(int)
    return df
```
